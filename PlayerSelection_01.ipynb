# Encoding categorical data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
# List of column names to be transformed
columns_to_encode = ['gender', 'country', 'strong_foot', 'dedication_level', 'coaching']
```

```python
# Apply LabelEncoder to the specified columns
for column in columns_to_encode:
    data[column] = le.fit_transform(data[column])
```

# split the Data

```python
features = ['age', 'gender', 'country',  'weight', 'fitness_rating', 'years_of_experience', 'no_of_disqualifications']
```

```python
x=data[features]
```

```python
y=data['selection']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Standardize features using StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Model Creation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Create a logistic regression model
model = LogisticRegression()

```

```python
# Train the model on the training data
model.fit(X_train_scaled, y_train)

```

```python
# Predictions on the testing data
predictions = model.predict(X_test_scaled)
```

```python

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)

```
